```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h)

# compute function
y = np.cos(x)

# compute vector of forward differences
forward_diff = np.diff(y)/h

# compute corresponding grid
x_diff = x[:-1:]

# compute exact solution
exact_solution = -np.sin(x_diff)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between
# numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
h = 0.1
x_values = np.array([0, 0.1, 0.2, 0.3, 0.4])
f_values = np.array([0.0000, 0.0819, 0.1341, 0.1646, 0.1797])

# Compute f'(0) using the O(h^2) finite difference approximation
f_prime_0 = (-3*f_values[0] + 4*f_values[1] - f_values[2]) / (2*h)
print(f"f'(0) ≈ {f_prime_0}")

# Compute vector of forward differences
forward_diff = np.diff(f_values) / h

# Compute corresponding grid
x_diff = x_values[:-1]

# Plot solution
plt.figure(figsize=(12, 8))
plt.plot(x_diff, forward_diff, '--', label='Forward difference approximation')
plt.scatter(x_diff, forward_diff)
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title('Finite Difference Approximation of the Derivative')
plt.legend()
plt.show()
```
